```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()

column_to_encoder=["workclass","education","maritalstatus","occupation","relationship","race","sex","native","native","Salary"]
for column in column_to_encoder:
  df_train[column]=LE.fit_transform(df_train[column])

column_to_encoder=["workclass","education","maritalstatus","occupation","relationship","race","sex","native","native","Salary"]
for column in column_to_encoder:
  df_test[column]=LE.fit_transform(df_test[column])
```

```python
X_train=df_train.iloc[:,0:13]
Y_train=df_train["Salary"]
```

```python
X_test=df_test.iloc[:,0:13]
Y_test=df_test["Salary"]
```

```python
from sklearn.naive_bayes import MultinomialNB
MNB=MultinomialNB()

MNB.fit(X_train,Y_train)
```

```python
Y_pred_train=MNB.predict(X_train)
Y_pred_test=MNB.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
ac1=accuracy_score(Y_train,Y_pred_train)
ac2=accuracy_score(Y_test,Y_pred_test)
print("Accuray Score Training Data:",ac1.round(4))
print("Accuracy Score Test Data:",ac2.round(4))
```
